```python
"""
ERNA ANALYSIS PIPELINE
--------------------------------
P1 = max amplitude in 2â€“7 ms window after stim onset
P2 = max amplitude in 7â€“9.9 ms window after stim onset
P1 amplitude = peak value
P2 amplitude = peak âˆ’ preceding trough
Always returns P1 and P2 when window contains samples.
"""

import numpy as np
import pandas as pd
from pathlib import Path
from neo.io import Spike2IO
from scipy.signal import butter, filtfilt, detrend as sp_detrend
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")

# =======================================================
# USER PATH
# =======================================================
AllSTNDataFolder = "/Users/arjunbalachandar/Desktop/University of Toronto/PhD/Y2/Courses/BME1500/STN Data"


# =======================================================
# PARAMETERS
# =======================================================

# Stim detection
V_THRESHOLD        = 3.0
V_HYSTERESIS       = 0.5
MIN_INTERSTIM_MS   = 3.0

# Artifact handling
ARTIFACT_GUARD_MS  = 0.8
EXTRA_GUARD_MS     = 0.4
MIN_SEG_MS         = 4.0

# Filtering + upsample
LPF_CUTOFF_HZ      = 700
LPF_ORDER          = 3
UPSAMPLE_FACTOR    = 10

# ERNA windows (relative to stim onset)
P1_LAT_MS = (2, 7)
P2_LAT_MS = (7, 9.9)

MAX_PULSE_PLOT = 60


# =======================================================
# UTIL
# =======================================================

def infer_freq_from_name(fname):
    up = fname.upper().replace(" ", "")
    if "0BL" in up:
        return 0
    import re
    m = re.search(r"(\d+)HZ", up)
    return int(m.group(1)) if m else None


def load_first_analog_signal(filepath: Path):
    reader = Spike2IO(filename=str(filepath))
    block  = reader.read_block()
    seg    = block.segments[0]
    if not seg.analogsignals:
        return None, None
    sig  = seg.analogsignals[0]
    data = sig.magnitude.flatten()
    fs   = float(sig.sampling_rate.magnitude)
    return data, fs


def detect_stim_onsets_threshold(raw, fs,
                                 stim_freq_hint_hz=None,
                                 v_threshold=3.0,
                                 hysteresis=0.5,
                                 min_interval_ms=3.0):
    """Threshold + hysteresis."""
    x = raw.astype(float)
    low = v_threshold - hysteresis
    above = x >= v_threshold
    below = x <= low

    onsets = []
    armed = bool(below[0])
    for i in range(1, len(x)):
        if not armed:
            if below[i]:
                armed = True
            continue
        if (not above[i-1]) and above[i]:
            onsets.append(i)
            armed = False
    onsets = np.array(onsets, dtype=int)

    # enforce min interstim interval
    if stim_freq_hint_hz and stim_freq_hint_hz > 0:
        period_s = 1.0 / stim_freq_hint_hz
        min_dist = int(0.8 * period_s * fs)
    else:
        min_dist = int((min_interval_ms/1000.0)*fs)

    if len(onsets) <= 1:
        return onsets

    keep = [onsets[0]]
    last = onsets[0]
    for idx in onsets[1:]:
        if idx-last >= min_dist:
            keep.append(idx)
            last = idx

    return np.array(keep, dtype=int)


def butter_lowpass(x, fs, fc=700, order=3):
    nyq = fs*0.5
    Wn = fc / nyq
    from scipy.signal import butter, filtfilt
    b,a = butter(order, Wn, btype="low")
    return filtfilt(b,a,x)


def upsample_timeseries(y, fs, factor=UPSAMPLE_FACTOR):
    if factor <= 1:
        return y, fs
    t    = np.arange(len(y)) / fs
    t_hi = np.linspace(t[0], t[-1], len(y)*factor)
    y_hi = CubicSpline(t, y)(t_hi)
    return y_hi, fs*factor


# =======================================================
# PEAK EXTRACTION BASED ON MAX AMPLITUDE WITHIN LAT WINDOW
# =======================================================

# =======================================================
# PEAK EXTRACTION â€” Updated
# P1 = max amp in 2â€“7 ms
# P2 = max amp in 7â€“9.9 ms
# (NO peak-to-trough subtraction)
# =======================================================

def compute_P1_P2(seg_u, fs_u, stim_offset_ms=0.0):
    """
    seg_u: filtered & upsampled segment
    fs_u : sampling rate after upsample
    stim_offset_ms: time shift (usually 0)
    
    Returns:
      P1_amp, P2_amp, P1_idx, P2_idx
    """

    # Build time axis aligned to stim onset
    time_ms = np.arange(len(seg_u)) / fs_u * 1000.0 - stim_offset_ms

    def get_window_idx(win):
        lo, hi = win
        mask = (time_ms >= lo) & (time_ms <= hi)
        return np.where(mask)[0]

    # ---- P1 ----
    idxs1 = get_window_idx(P1_LAT_MS)
    P1_amp = np.nan
    P1_idx = None
    if idxs1.size > 0:
        seg_sub = seg_u[idxs1]
        loc = np.argmax(seg_sub)
        P1_idx = idxs1[loc]
        P1_amp = seg_u[P1_idx]     # peak only

    # ---- P2 ----
    idxs2 = get_window_idx(P2_LAT_MS)
    P2_amp = np.nan
    P2_idx = None
    if idxs2.size > 0:
        seg_sub = seg_u[idxs2]
        loc = np.argmax(seg_sub)
        P2_idx = idxs2[loc]
        P2_amp = seg_u[P2_idx]     # peak only

    return P1_amp, P2_amp, P1_idx, P2_idx


# =======================================================
# MAIN PROCESSING
# =======================================================

root = Path(AllSTNDataFolder)
trial_dirs = sorted([d for d in root.iterdir() if d.is_dir()])
rows = []

for tr in trial_dirs:
    for f in sorted(tr.glob("*.smr")):

        freq = infer_freq_from_name(f.name)
        if freq is None or freq == 0:
            continue

        data, fs = load_first_analog_signal(f)
        if data is None:
            continue

        onsets = detect_stim_onsets_threshold(
            data, fs,
            stim_freq_hint_hz=freq,
            v_threshold=V_THRESHOLD,
            hysteresis=V_HYSTERESIS,
            min_interval_ms=MIN_INTERSTIM_MS
        )
        if len(onsets) < 2:
            continue

        P1_a=[]; P2_a=[]; pIdx=[]
        P1_i=[]; P2_i=[]

        for p_idx in range(len(onsets)-1):
            s0, s1 = onsets[p_idx], onsets[p_idx+1]

            # segment between pulses
            seg = data[s0:s1].astype(float)

            # detrend + LPF
            seg = sp_detrend(seg, type="linear")
            seg = butter_lowpass(seg, fs, fc=LPF_CUTOFF_HZ, order=LPF_ORDER)

            # upsample
            seg_u, fs_u = upsample_timeseries(seg, fs, factor=UPSAMPLE_FACTOR)

            # compute peaks relative to stim onset = time 0
            P1_amp, P2_amp, P1_idx, P2_idx = compute_P1_P2(
                seg_u, fs_u, stim_offset_ms=0.0
            )

            P1_a.append(P1_amp)
            P2_a.append(P2_amp)
            P1_i.append(P1_idx)
            P2_i.append(P2_idx)
            pIdx.append(p_idx)

        # normalize within file
        if len(P1_a)>0:
            P1_0 = P1_a[0] if np.isfinite(P1_a[0]) and P1_a[0]!=0 else np.nan
            P2_0 = P2_a[0] if np.isfinite(P2_a[0]) and P2_a[0]!=0 else np.nan

            for (pi,pa,pb,i1,i2) in zip(pIdx,P1_a,P2_a,P1_i,P2_i):
                P1_rel = pa/P1_0 if np.isfinite(pa) and np.isfinite(P1_0) and P1_0!=0 else np.nan
                P2_rel = pb/P2_0 if np.isfinite(pb) and np.isfinite(P2_0) and P2_0!=0 else np.nan

                rows.append(dict(
                    Trial=tr.name,
                    File=f.name,
                    StimFreq_Hz=freq,
                    PulseIndex=pi,
                    P1_abs=pa,
                    P2_abs=pb,
                    P1_rel=P1_rel,
                    P2_rel=P2_rel,
                    P1_idx=i1,
                    P2_idx=i2
                ))


df = pd.DataFrame(rows)
df.to_csv("ERNA_latency_relative_by_pulse.csv", index=False)
print("\nSaved â†’ ERNA_latency_relative_by_pulse.csv")
print(df.head())

# =======================================================
# PLOTTING â€” RELATIVE ERNA (50 & 100 Hz, OUTLIERS REMOVED)
# =======================================================

MAX_PULSE_PLOT = 50

if len(df):

    # Clean numeric stim frequency
    df = df.copy()
    df["StimFreq_Num"] = pd.to_numeric(
        df["StimFreq_Hz"].astype(str).str.extract(r"(\d+)")[0],
        errors="coerce"
    )

    # Limit dataframe: first N pulses + only 50 & 100 Hz
    df_sel = df[
        (df["PulseIndex"] < MAX_PULSE_PLOT) &
        (df["StimFreq_Num"].isin([50, 100]))
    ].copy()

    print("Before outlier removal:", len(df_sel))

    # ====================================================
    # IQR-based outlier removal
    # Remove rows where P1_rel or P2_rel are extreme
    # ====================================================
    def remove_outliers_IQR(series, k=3.0):
        """Return boolean mask where values NOT outliers."""
        q1 = series.quantile(0.25)
        q3 = series.quantile(0.75)
        iqr = q3 - q1
        lower = q1 - k * iqr
        upper = q3 + k * iqr
        return (series >= lower) & (series <= upper)

    # Remove outliers from P1_rel and P2_rel separately
    mask_P1 = remove_outliers_IQR(df_sel["P1_rel"])
    mask_P2 = remove_outliers_IQR(df_sel["P2_rel"])
    df_filt = df_sel[mask_P1 & mask_P2].copy()

    print("After outlier removal:", len(df_filt))

    # Confirm remaining freqs
    print("Plotting frequencies:", sorted(df_filt["StimFreq_Num"].unique()))

    # ---------------------------------------------------
    # P1 â€” Relative amplitude vs pulse index
    # ---------------------------------------------------
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df_filt,
        x="PulseIndex",
        y="P1_rel",
        hue="StimFreq_Num",
        estimator="mean",
        ci=95,
        marker="o",
        hue_order=[50,100]
    )
    plt.title("Relative P1 Amplitude vs Pulse Index (50 & 100 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative P1 Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---------------------------------------------------
    # P2 â€” Relative amplitude vs pulse index
    # ---------------------------------------------------
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df_filt,
        x="PulseIndex",
        y="P2_rel",
        hue="StimFreq_Num",
        estimator="mean",
        ci=95,
        marker="o",
        hue_order=[50,100]
    )
    plt.title("Relative P2 Amplitude vs Pulse Index (50 & 100 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative P2 Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ======================================================
# Count number of trials contributing to each curve
# ======================================================

trial_counts = (
    df_filt.groupby("StimFreq_Num")["Trial"]
    .nunique()
    .reset_index()
    .rename(columns={"Trial": "NumTrials"})
)

print("\nNumber of unique trials represented:\n")
print(trial_counts.to_string(index=False))


# =======================================================
# QC PLOTS â€” P1/P2 markers on raw trace (first 10 files)
# =======================================================

N_QC_TRIALS = 10
WINDOW_MS   = 50

plot_traces = False  # set to True to enable raw trace plotting

def qc_plot_file(df_sub, data, fs, onsets, fs_u, factor, title=""):

    win_raw = int(WINDOW_MS/1000 * fs)

    df_sub = df_sub.sort_values("PulseIndex")
    nplot = min(len(onsets)-1, 10)

    fig, axes = plt.subplots(nplot, 1, figsize=(10,2*nplot), sharex=True)
    if nplot == 1:
        axes = [axes]

    for i in range(nplot):
        s0 = onsets[i]
        start = s0
        end   = s0 + win_raw
        if end > len(data):
            end = len(data)

        raw_seg = data[start:end]
        time_raw_ms = np.arange(len(raw_seg)) / fs * 1000

        ax = axes[i]
        ax.plot(time_raw_ms, raw_seg, color="black", lw=1)

        pulse = df_sub[df_sub["PulseIndex"] == i]
        if len(pulse):
            P1_idx = pulse["P1_idx"].values[0]
            P2_idx = pulse["P2_idx"].values[0]

            if np.isfinite(P1_idx):
                tP1 = (P1_idx / fs_u)*1000
                raw_i = min(int(P1_idx/factor), len(raw_seg)-1)
                ax.scatter(tP1, raw_seg[raw_i], color="red", s=30, label="P1" if i==0 else "")

            if np.isfinite(P2_idx):
                tP2 = (P2_idx / fs_u)*1000
                raw_i = min(int(P2_idx/factor), len(raw_seg)-1)
                ax.scatter(tP2, raw_seg[raw_i], color="blue", s=30, label="P2" if i==0 else "")

        if i == 0:
            ax.legend(loc="upper right")

        ax.set_ylabel(f"Pulse {i}")
        ax.grid(True)

    axes[-1].set_xlabel("Time (ms)")
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()


files_seen = 0
for tr in trial_dirs:
    for f in sorted(tr.glob("*.smr")):

        freq = infer_freq_from_name(f.name)
        if freq is None or freq == 0 or freq == 40:
            continue

        data, fs = load_first_analog_signal(f)
        if data is None:
            continue

        onsets = detect_stim_onsets_threshold(
            data, fs,
            stim_freq_hint_hz=freq,
            v_threshold=V_THRESHOLD,
            hysteresis=V_HYSTERESIS,
            min_interval_ms=MIN_INTERSTIM_MS
        )
        if len(onsets) < 2:
            continue

        df_sub = df[df["File"] == f.name]
        if len(df_sub) == 0:
            continue

        fs_u = fs * UPSAMPLE_FACTOR
        
        if plot_traces:
            qc_plot_file(
                df_sub, data, fs, onsets,
                fs_u=fs_u, factor=UPSAMPLE_FACTOR,
                title=f"{tr.name} â€” {f.name} ({freq} Hz)"
            )

        files_seen += 1
        if files_seen >= N_QC_TRIALS:
            break
    if files_seen >= N_QC_TRIALS:
        break


# =======================================================
# NEW PLOTS â€” 100 Hz ONLY
# =======================================================

df60 = df[df["PulseIndex"] < MAX_PULSE_PLOT].copy()
df100 = df60[df60["StimFreq_Hz"] == 100]

if len(df100):

    # ---- P1 only, 100 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df100,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )
    plt.title("Relative P1 amplitude vs Pulse Index (100 Hz)")
    #change y-axis label
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- P2 only, 100 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df100,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative P2 Amplitude vs Pulse Index (100 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- P2 only, 100 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df100,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )

    sns.lineplot(
        data=df100,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative Amplitude vs Stimulus Number (100 Hz)")
    plt.xlabel("Stimulus number")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("No 100 Hz data available for plotting.")

# =======================================================
# NEW PLOTS â€” 50 Hz ONLY
# =======================================================

# Select first 60 pulses & freq = 50 Hz
df50 = df[df["PulseIndex"] < MAX_PULSE_PLOT]
df50 = df50[df50["StimFreq_Num"] == 50]

if len(df50):

    # ---- P1 only, 50 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df50,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )
    plt.title("Relative P1 Amplitude vs Pulse Index (50 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- P2 only, 50 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df50,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative P2 Amplitude vs Pulse Index (50 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- Combined P1 & P2, 50 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df50,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )
    sns.lineplot(
        data=df50,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative Amplitude vs Stimulus Number (50 Hz)")
    plt.xlabel("Stimulus Number")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("No 50 Hz data available for plotting.")

# =======================================================
# NEW PLOTS â€” 50 Hz ONLY
# =======================================================

# Select first 60 pulses & freq = 50 Hz
df30 = df[df["PulseIndex"] < MAX_PULSE_PLOT]
df30 = df30[df30["StimFreq_Num"] == 30]

if len(df30):

    # ---- P1 only, 30 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df30,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )
    plt.title("Relative P1 Amplitude vs Pulse Index (30 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- P2 only, 30 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df30,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative P2 Amplitude vs Pulse Index (30 Hz)")
    plt.xlabel("Pulse Index")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ---- Combined P1 & P2, 30 Hz ----
    plt.figure(figsize=(8,6))
    sns.lineplot(
        data=df30,
        x="PulseIndex",
        y="P1_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="red"
    )
    sns.lineplot(
        data=df30,
        x="PulseIndex",
        y="P2_rel",
        estimator="mean",
        ci=95,
        marker="o",
        color="blue"
    )
    plt.title("Relative Amplitude vs Stimulus Number (30 Hz)")
    plt.xlabel("Stimulus Number")
    plt.ylabel("Relative Amplitude [a.u.]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("No 30 Hz data available for plotting.")

```

```python
# ======================================================
# RELATIVE AMPLITUDE: MIXED EFFECTS MODEL â€” FIRST 60 PULSES ONLY
# RelAmp ~ PeakType * PulseIndex + (1 | Trial)
# ======================================================

import pandas as pd
import statsmodels.formula.api as smf

MAX_PULSES = 60   # limit to first 60

# ---- Step 0 â€” restrict df to first 60 pulses ----
df_sub = df[df["PulseIndex"] < MAX_PULSES].copy()

# ---- Step 1 â€” reshape to long format ----
df_long = pd.melt(
    df_sub,
    id_vars=["Trial", "File", "StimFreq_Hz", "PulseIndex"],
    value_vars=["P1_rel", "P2_rel"],
    var_name="PeakType",
    value_name="RelAmp"
)

# Rename peak label
df_long["PeakType"] = df_long["PeakType"].map({
    "P1_rel": "P1",
    "P2_rel": "P2"
})

# Drop rows missing amplitude
df_long = df_long.dropna(subset=["RelAmp"])

# Fix types
df_long["PulseIndex"] = df_long["PulseIndex"].astype(float)
df_long["PeakType"]   = df_long["PeakType"].astype("category")
df_long["Trial"]      = df_long["Trial"].astype("category")

print(" First 60 pulses â€” Long format:")
print(df_long.head())

# ---- OPTIONAL: center pulse index ----
df_long["PulseIndex_c"] = df_long["PulseIndex"] - df_long["PulseIndex"].mean()

# ---- Step 2 â€” Fit mixed model ----
# NOTE: formula still uses PulseIndex (not PulseIndex_c),
#       but you may swap if preferred.
model60 = smf.mixedlm(
    "RelAmp ~ PeakType * PulseIndex",
    data=df_long,
    groups=df_long["Trial"]
).fit()

print("\n Mixed-effects model (first 60 pulses) fitted successfully\n")
print(model60.summary())
```

```python
# =======================================================
#   PER-TRIAL SUMMARY:
#   Mean P1 / P2  +  time to first spike post-train
# =======================================================

import numpy as np
import pandas as pd
from pathlib import Path

# -------------------------
# Helper: simple spike finder
# -------------------------
def find_first_spike_time(raw, fs, last_stim_idx, threshold=None):
    """
    Find time (ms) from last stimulus to first threshold-crossing spike.
    Spike = raw < threshold (neg deflection).
    Returns np.nan if no spike occurs.
    """

    segment = raw[last_stim_idx:]
    if threshold is None:
        # Default: negative spike threshold
        mu  = np.mean(segment)
        sd  = np.std(segment)
        threshold = mu - 2.5 * sd   # i.e. significant negative crossing

    # Detect threshold-crossing samples
    crossing = np.where(segment < threshold)[0]
    if len(crossing) == 0:
        return np.nan

    # First spike sample
    spike_sample = crossing[0]
    time_ms = spike_sample / fs * 1000.0
    return time_ms


# =======================================================
# Build per-trial table of mean P1/P2 & time-to-first-spike
# =======================================================

summary_rows = []

df_copy = df.copy()

# Clean numeric stim freq
df_copy["StimFreq_Num"] = pd.to_numeric(
    df_copy["StimFreq_Hz"].astype(str).str.extract(r"(\d+)")[0],
    errors="coerce"
)

# Loop by trial + file
for (trial, fname, stimF), sub in df_copy.groupby(["Trial","File","StimFreq_Num"]):

    # --- Mean P1/P2 ---
    mean_P1 = sub["P1_abs"].mean()
    mean_P2 = sub["P2_abs"].mean()

    # --- Load raw & detect stim ---
    fpath = Path(AllSTNDataFolder) / trial / fname
    data, fs = load_first_analog_signal(fpath)
    if data is None:
        continue

    # Detect stim pulses
    onsets = detect_stim_onsets_threshold(
        data, fs,
        stim_freq_hint_hz=stimF,
        v_threshold=V_THRESHOLD,
        hysteresis=V_HYSTERESIS,
        min_interval_ms=MIN_INTERSTIM_MS
    )
    if len(onsets) == 0:
        continue

    last_stim_idx = onsets[-1]

    # --- Find first spike after last pulse ---
    t_spk = find_first_spike_time(data, fs, last_stim_idx)

    summary_rows.append(dict(
        Trial       = trial,
        File        = fname,
        StimFreq    = stimF,
        MeanP1_abs  = mean_P1,
        MeanP2_abs  = mean_P2,
        TimeFirstSpike_ms = t_spk
    ))

summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv("ERNA_trial_summary.csv", index=False)
print("\nâœ… Saved â†’ ERNA_trial_summary.csv")
print(summary_df.head())
```
